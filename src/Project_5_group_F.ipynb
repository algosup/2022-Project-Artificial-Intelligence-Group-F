# Setup the datasets

Declare the variables 

```python
train = "train"
test = "test"

eng = "english"
fr = "french"

categories = [train, test]
languages = [eng, fr]

original_dataset_dir = {}

original_dataset_dir[eng] = "/datasets/english/"
original_dataset_dir[fr] = "/datasets/french/"

dataset_root_path = '/content/data/'
target_dataset_dir = '/content/data/'
train_path = dataset_root_path + train


num_files_to_take = 20000 # number of files to take from each language
train_rate = 0.8 # 80% train, 20% test


batch_size = 128
image_width = 500
image_height = 128

validation_split = 0.1
learning_rate = 0.045

sample_rate = 8000

num_classes = len(languages)

model_file = dataset_root_path + 'model.h5'


augment_data_factor = 1.0 # Set-up how much of the training data should be augmented
```

Checks the paths to the datasets

```python
for lang in languages:
    if not os.path.isdir(original_dataset_dir[lang]):
        raise
    for category in categories:
        if not os.path.isdir(target_dataset_dir + category + '/' + lang):
            raise

for lang in languages:
    if not os.path.isfile(original_dataset_dir[lang] + 'validated.tsv'):
        raise
    if not os.path.isdir(original_dataset_dir[lang] + 'clips'):
        raise
```

Load the data

```python
def copy_audio_files_for_language(lang):

    print("Copying audio files for language: " + lang + "<PROCESSING>")

    df = pd.read_csv(original_dataset_dir[lang] + 'validated.tsv', sep='\t')
    all_filenames = df['path'].tolist()
    shuffle(all_filenames)

    counter = 0

    category = train

    for filename in all_filenames:
        file = original_dataset_dir[lang] + 'clips/' + filename
        try:
            audio_segment, sample_rate = lr.load(file)
            if np.count_nonzero(audio_segment) == 0:
                raise Exception("Empty audio file")
            if audio_segment.ndim != 1:
                raise Exception("Audio file is not compatible" + str(audio_segment.ndim))
            duration_sec = lr.core.get_duration(audio_segment, sr=sample_rate)
        except Exception as e:
            print("ERROR in file: " + file + " " + str(e))
            continue


        
        if 4.0 < duration_sec < 10.0: # only take files with a duration between 4 and 10 seconds
            copy2(file, target_dataset_dir + category + '/' + lang + '/' + filename)
            counter += 1


        if counter == int(num_files_to_take * train_rate):
            category = test
        if counter == num_files_to_take:
            break
```

```python
warnings.simplefilter('ignore', UserWarning)

copy_audio_files_for_language(fr)

warnings.simplefilter('default', UserWarning)
```

# Data augmentation

Load the data

```python
def load_audio_files(audio_file_path):
    warnings.simplefilter('ignore', UserWarning)

    audio_segment, sample_rate = lr.load(audio_file_path)
    return audio_segment, sample_rate

    warnings.simplefilter('default', UserWarning)
```

Data Augmentation

```python
def add_noise(audio_segment, gain):
    num_samples = audio_segment.shape[0]
    noise = gain * np.random.normal(size=num_samples)
    return audio_segment + noise
```

All availables data :

```python
audio_files = {}

for lang in languages:
    for category in categories:
        audio_files[lang + "." + category] = glob(dataset_root_path + category + '/' + lang + '/*.mp3')

print(audio_files.keys())
```

Apply the functions

```python
def augment_audio_file_with_noise(audio_file_path):
    audio_segment, sample_rate = load_audio_files(audio_file_path)
    audio_segment_with_noise = add_noise(audio_segment=0.005)
    audio_file_path_without_extension = os.path.splitext(audio_file_path)[0]
    augmented_audio_file_path = audio_file_path_without_extension + '_augmented.wav'
    sf.write(augmented_audio_file_path, audio_segment_with_noise, sample_rate)
```

```python
for lang in languages:
    category = train

    all_audio_files = audio_files[lang + "." + category]

    num_files = int(len(all_audio_files) * augment_data_factor)

    for i in range(num_files):
        if i % (num_files / 10) == 0:
            print("Still processing " + lang + ' ' + category + ' ' + str(i) + '/' + str(num_files))
        augment_audio_file_with_noise(all_audio_files[i])
```

 # Data processing

Load the data

```python
def load_audio_file(audio_file_path):
    warnings.simplefilter('ignore', UserWarning)
    
    audio_segment, _ = lr.load(audio_file_path, sr=sample_rate)
    return audio_segment

    warnings.simplefilter('default', UserWarning)
```

Set all the sound to 10 seconds

```python
def fix_audio_segment_to_10_seconds(audio_segment):
    target_len = 10 * sample_rate
    audio_segment = np.concatenate([audio_segment]*3, axis=0)
    audio_segment = audio_segment[0:target_len]

    return audio_segment
```

Generate the spectrogram from the samples

```python
def spectrogram(audio_segment): 
    h1 = audio_segment.shape[0] // image_width
    spec = lr.feature.melspectrogram(audio_segment, n_mels=image_height, hop_length=int(h1))


    image = lr.core.power_to_db(spec)

    image_np = np.asmatrix(image)

    image_np_scaled_temp = (image_np - np.min(image_np))

    image_np_scaled = image_np_scaled_temp / np.max(image_np_scaled_temp)

    return image_np_scaled[:, 0:image_width]
```

Convert data to integer 

```python
def to_integer(image_float):
    image_float_255 = image_float * 255

    # Convert to uint8
    image_int = image_float_255.astype(np.uint8)

    return image_int
```

Convert the audio files to a spectrogram and store them as .png

```python
def audio_to_image_file(audio_file):
    out_image_file = audio_file + '.png'
    audio = load_audio_file(audio_file)
    audio_fixed = fix_audio_segment_to_10_seconds(audio)
    if np.count_nonzero(audio_fixed) != 0:
        spectro = spectrogram(audio_fixed)
        spectro_int = to_integer(spectro)
        imageio.imwrite(out_image_file, spectro_int)
    else:
        print("Audio files is empty")
```

# Train and test the dataset

```python
all_files = glob(train_path + '/*/*.png')

num_validation = len(all_files) * validation_split
num_training = len(all_files) - num_validation

validation_steps = int(num_validation / batch_size)
training_steps = int(num_training / batch_size)

print('steps per epoch:', str(training_steps))
print('validation steps:', str(validation_steps))
```

Training batch generator function

```python
image_data_generator = ImageDataGenerator(rescale=1./255, validation_split=validation_split)
train_generator = image_data_generator.flow_from_directory(train_path, batch_size=batch_size, class_mode='categorical', target_size=(image_height, image_width), color_mode='grayscale', subset='training')
validation_generator = image_data_generator.flow_from_directory(train_path, batch_size=batch_size, class_mode='categorical', target_size=(image_height, image_width), color_mode='grayscale', subset='validation')
```

Model definition

```python
img_input = Input(shape=(image_height, image_width, 1))

img_conc = Concatenate(axis=3, name='concatenate')([img_input, img_input, img_input])

model = InceptionV3(input_tensor=img_conc, weights=None, include_top=True, classes=2)

model.Summary()
```

Compiling the model

```python
model.compile(optimizer=RMSprop(lr=learning_rate, clipvalue=2.0), loss='categorical_crossentropy', metrics=['accuracy'])
```

Training the model

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=10, mode='max', restore_best_weights=True)
```

Learning rate schedule

```python
def step_decay(epoch, lr):
    lrate = lr * math.pow(drop, math.floor((1+epoch)/epochs_drop))
    drop = 0.94
    epochs_drop = 2.0
    return lrate

learning_rate_decay = LearningRateScheduler(step_decay, verbose=1)
```

```python
Training = model.fit_generator(train_generator, steps_per_epoch=training_steps, epochs=100, validation_data=validation_generator, validation_steps=validation_steps, callbacks=[early_stopping, learning_rate_decay])
```

Save the model 

```python
model.save(model_file)
```

# Display the result

```python
plt.plot(Training.history['acc'])
plt.plot(Training.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
